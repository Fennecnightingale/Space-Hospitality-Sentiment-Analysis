```python
def vectorizer_and_stacker(dataframe):
    """
    Pass in a DataFrame & this function will standardize, tdidf vectorize, and decrease dimensions
    Returns completed csr_matrix 
    """
    dataframe = dataframe[['favorite_count', 'repost_count', 'text']].dropna()
    #turning our other x variables into a matrix so we can combine all our features into one
    haystack = MaxAbsScaler().fit_transform(csr_matrix(dataframe[['favorite_count', 'repost_count']]))
    # create the TDIDF transform
    vect = TfidfVectorizer().fit_transform(dataframe['text'])
    print('tfidf vectorizing complete')
    # doing some PCA to reduce our dimensions to make this faster and easier to work with
    svd = TruncatedSVD(n_components=100, n_iter=25000, random_state=0, algorithm='arpack').fit(vect)
    # splitting it so that when it is transformed into an array it doesn't need to store it as one large array and 
    # hopefully won't cause memory errors. start with a list to put slices of our matrix into.
    splits = []
    shape = vect.shape[1]
    tbs = vect 
    # since you cant directly index a matrix the easiest/fastest/computationally cheapest way is train_test
    # chose 30000 since my computer can definitely handle a chunk of that size 
    while shape > 30000:
        # the train size/shape make it so our train will contain 30k and our test will contain everything else
        train_size = 30000/shape
        # we continually update shape so we know when to stop
        shape = shape - 30000
        # for our first run through we take the whole matrix, then we update tbs to be everything 
        split1, split2 = train_test_split(tbs, random_state=0, train_size=train_size)
        # except the chunk we just removed, so all of our leftover rows (not in the 30k) are in split2
        tbs = split2
        # put the chunk of our matrix into a list to iterate over later 
        splits.append(split1)
    # append the remainder (whatever was left < 30k) to the end of our list 
    splits.append(tbs)
    # start a new matrix by fit transforming our first chunk
    init = csr_matrix(svd.fit_transform(splits[0]))
    # start a loop that will fit transform and stack our new transformed matricies together
    for item in splits[1:]:
        # fit transform will return it as a dense array, so we also need to convert it back here 
        init = vstack([init, csr_matrix(svd.fit_transform(item))])
    # combine our text matrix (init) & our repost/favorite matrix (haystack)
    haystack = hstack([init, haystack])
    print('dimensionality reduction complete')
    # stacking our matricies into a single matrix to model off of
    print('matrix complete')
    return haystack
```

```python
def clusterizer(dataframe, haystack, num_clusters, n_init, max_iter, nredo, niter):
  print('clusterizer initalizing')
  d = coo_matrix.toarray(stack).astype(np.float32)
  n_init = n_init
  max_iter = max_iter
  kmeans = faiss.Kmeans(d= d.shape[1],
                        k=num_clusters,
                        nredo = nredo,
                        verbose = True,  
                        update_index = True,
                        max_points_per_centroid = len(dataframe)//num_clusters,
                        min_points_per_centroid = len(dataframe)//num_clusters,
                        niter = niter)
  kmeans.train(d)

  fit_predict = kmeans.index.search(d, 1)[1]
  dataframe['cluster'] = fit_predict
  return kmeans, fit_predict, kmeans.obj, kmeans.iteration_stats
```

```python
def make_silhouette(haystack, cluster, array, csv, fit_predict, inertia_):
    print('gathering cluster labels')
    cluster_labels = np.unique(fit_predict)
    n_clusters = cluster_labels.shape[0]
    print('silhouette for ' + csv[12:14].replace('_', ' ') + ' words ' + str(n_clusters) + ' clusters ' + str(round(inertia_, 2)) + ' inertia')
    silhouette_vals = silhouette_samples(haystack, fit_predict, metric='l2', n_jobs=-1)
    y_ax_lower, y_ax_upper = 0, 0
    yticks = []
    for i, c in enumerate(cluster_labels):
        c_silhouette_vals = silhouette_vals[fit_predict == c]
        c_silhouette_vals.sort()
        y_ax_upper += len(c_silhouette_vals)
        color = cm.jet(float(i) / n_clusters)
        plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, 
                 edgecolor='none', color=color)
        yticks.append((y_ax_lower + y_ax_upper) / 2.)
        y_ax_lower += len(c_silhouette_vals)
    silhouette_avg = np.mean(silhouette_vals)
    plt.axvline(silhouette_avg, color="red", linestyle="--") 
    plt.yticks(yticks, cluster_labels + 1)
    plt.ylabel('Cluster')
    plt.xlabel('Silhouette coefficient')
    plt.savefig(str(round(inertia_, 0)) + 'k_' + csv[12:] + '_' + str(n_clusters) + '_clst.png', dpi=300)
    plt.tight_layout()
    plt.show()
```

```python
def plt_elbow(haystack, dataframe, n_init, max_iter, nredo, niter):
    MAX_K = 5
    K = range(1,MAX_K+1)
    inertias = np.zeros(MAX_K)
    diff = np.zeros(MAX_K)
    diff2 = np.zeros(MAX_K)
    diff3 = np.zeros(MAX_K)
    imb_mean = []
    for k in K:
        d = coo_matrix.toarray(haystack).astype(np.float32)
        n_init = n_init
        max_iter = max_iter
        kmeans = faiss.Kmeans(d= d.shape[1],
                              k=k,
                              nredo = nredo,  
                              seed = 0,
                              verbose = True,
                              # update_index = True,
                              max_points_per_centroid = (len(dataframe)//k) + 1,
                              min_points_per_centroid = (len(dataframe)//k) - 1, 
                              niter = niter)
        kmeans.train(d)    
        inertias[k - 1] = kmeans.obj[-1]
        avg = []
        for item in kmeans.iteration_stats:
            avg.append(item['imbalance_factor'])
        print('Cluster: ', kmeans.iteration_stats[0]['nsplit'],  'Inertia: ', round(kmeans.obj[-1], 2), 
              'Imbalance Mean: ', round(np.mean(avg), 2), 'Imbalance Median: ', round(np.median(avg), 2))
        imb_mean.append(np.mean(avg))
        # first difference    
        if k > 1:
            diff[k - 1] = inertias[k - 1] - inertias[k - 2]
        # second difference
        if k > 2:
            diff2[k - 1] = diff[k - 1] - diff[k - 2]
        # third difference
        if k > 3:
            diff3[k - 1] = diff2[k - 1] - diff2[k - 2]
    elbow = np.argmin(diff3[3:]) + 3
    print("Elbow "+str(elbow))
    fit_predict = kmeans.index.search(d, 1)[1]
    # label our results in our dataframe 
    dataframe['cluster'] = fit_predict
    plt.plot(K, inertias, "b*-")
    plt.plot(K[elbow], inertias[elbow], marker='o', markersize=12,
             markeredgewidth=2, markeredgecolor='r', markerfacecolor=None)
    plt.ylabel("Inertia")
    plt.xlabel("K")
    plt.show()
    return K[elbow], imb_mean[K[elbow]]
```

```python
# generate tf-idf vectorization (use sklearn's TfidfVectorizer) for our data
def tfidf(X, y, stopwords_list): 
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
    vectorizer = TfidfVectorizer(stop_words=stopwords_list)
    tf_idf_train = vectorizer.fit_transform(X_train)
    tf_idf_test = vectorizer.transform(X_test)
    return tf_idf_train, tf_idf_test, y_train, y_test, vectorizer
```

```python
#function that calculates the inverse document frequency(IDF) of each word in our collection
def get_idf(class_, dataframe, stopwords_list):
    docs = dataframe[dataframe.cluster==class_].text
    class_dict = {} 
    for doc in docs:
        words = set(doc.split())
        for word in words:
            if word.lower() not in stopwords_list: 
                class_dict[word.lower()] = class_dict.get(word.lower(), 0) + 1
    idf_df = pd.DataFrame.from_dict(class_dict, orient='index')
    idf_df.columns = ['IDF']
    idf_df.IDF = np.log(len(docs)/idf_df.IDF)
    idf_df = idf_df.sort_values(by="IDF", ascending=True)
    return idf_df.head(40)
```
